```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[4,5,10,4,10,8,3,2,7,3]
y=[23,27,22,30,21,23,27,29,30,25]
```

```python
data=list(zip(x,y))
data
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)
model.fit(data)
```

```python
plt.scatter(x,y,c=model.labels_)
plt.show()
```
